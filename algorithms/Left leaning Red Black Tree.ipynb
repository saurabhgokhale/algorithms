# Left Leaning Red Black Tree

Left leaning red black tree is a actual implementation of 2-3 tree. It implements 2-3 tree as a BST.

This data structure was invented at Xerox park. Xerox invented color lazer printing and at that time, red color looked good. :)  

## 2-3 tree
2-3 tree is a tree similar to a binary tree where a node can have 3 children instead of max 2 like in case of binary tree.

The biggest advantage of 2-3 tree is, it maintains symmetric order and perfect balance. That is **Every path from root to null link has same length**

In worst case if every node has 2 children (2 nodes) then its Log2(N) 
in best case if every node has 3 children then its Log3(N) ==> 0.631 log2(N)

## Idea of the Left leaning RB Tree

We use **internal** left leaning links as **glue** for 3 nodes

![image](https://user-images.githubusercontent.com/2688478/33696503-64015164-dab7-11e7-8542-7010da29ebec.png)

### Properties of RB tree
1. No node has two red links connected to it.
2. Every path from root to null link has the same number of black links (Perfect black balance)
3. Red links lean left

![image](https://user-images.githubusercontent.com/2688478/33696546-9d1b2146-dab7-11e7-8714-42c6f9aba309.png)

### Search in RB

Search in Red-Black BST is the same as for elementary BST (We ignore the color of the link). But the performance of this search is much better as the tree is much better balanced.

![image](https://user-images.githubusercontent.com/2688478/33696574-baa08bac-dab7-11e7-8237-913d71916c8d.png)

### Representation of RED link

We add color property to the node itself. We assume that this color is the color of the link it connects with the parent.

![image](https://user-images.githubusercontent.com/2688478/33696655-22f7e1dc-dab8-11e7-9a09-133337b36180.png)

In the method isRed(Node x), if the link to the child is null then its not a red link and we return false. If the link was red it would mean incomplete 3 node which is not correct.
