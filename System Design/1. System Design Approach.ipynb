# System Design Approach


## ABCD of system design

[A]sk Good Questions
- Talk to interviewer to figure out which **Features** he/she cares about.
  - Feature set needs to be specific / small
  - We do deep into the feature set
- How much the system should scale 
  - **Request per Second** - How many requests per seconds needs to be handled
  - **Latency** - How much latency is ok for the system to respond
  - **Storage** - How much data needs to be stored in the database
  
Dont use [B]uzzwords 
- Dont talk about topics that you have just heard of and does not have indepth knowledge of.

[C]lear and Organized Thinking
- Before jumping into minor details of the problems, first design 20 thousand view of the application.
- Make sure we design all the APIs
- Define Actors for the application

[D]rive Discussion
- 80-20 Rule: We speak 80% of the time, interviewer speaks 20% of the time, we drive the discussion and the solution.


## Basic Features of System Design

- **Features** - Define minimum viable products. Keeps some features in the design and exclude some features. Based on the type of system some of the features like security can be ignored / kept out of the design. But if the design is about Authentication then security is a central piece of the system.

- **Define APIs**

- **Availability**

- **Latency / Performance**

- **Scalability**

- **Durability** - To make sure data is secure and consistent

- **Class Diagram**

- **Security and Privacy**

- **Cost Effective solution**


## Important Concepts for the Design questions

- Vertical vs Horizontal Scalability
- CAP theorem
- Partitioning / Sharding Data
- Optimistic vs Pessimistic locking
- Strong vs Eventual Consistency
- Relational DB vs No SQL
- Type of NO SQL
  - Key / Value
  - Wide column
  - Document based
  - Graph based
- Caching
- Data center / Racks / Hots
- CPU / memory / HD / Network Bandwidth
- Random vs Sequential read/write on disk
- HTTP vs HTTP2 vs websockets
- TCP / IP model
- IPv4 vs IPv6
- TCP vs UDP
- DNS lookup
- HTTPS & TLS
- Public key / Private key - Symmetric vs Asymetric key encryption
- LB - L4 vs L7
- CDN & Edge
- Bloom filters & Count min Sketch
- Paxos - Consensus over distributed hosts
  - Leader election
- Design patterns and Object oriented design
- Virtual machines & containers
- Publish - Subscribe messaging or queue
- Map Reduce
- Multi threading, concurrency, locks, synchronization, CAS

## Actual Implementations of the above concepts
- Cassandra
  - Uses eventual / strong consistency. Uses algorithm to shard data. Gossiping to keep cluster of servers.
- Mongodb / Couchbase
  - ACID properties at document level
- MySQL / Postgres
  - Traditional ACID properties, secondary indexes
- Memcache / Redis
  - Fast key value storage
  - can be stored as cluster
  - Note: Distributed cache can never be the **source of truth** and then can only store **limited** amount of data, limited by amount of memory on the machine.
- Zookeeper
- Kafka
  - fault tolerent queue
- Nginx / HAProxy
- Solr / Elastic Search
- Blob store (like S3)
- Docker
  - Swarm
  - Kubernates
- Hadoop / Spark
  - HDFS
  - Avro
  - Parquet
